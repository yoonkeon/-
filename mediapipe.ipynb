```python
import cv2
import mediapipe as mp
import time
    
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

cap = cv2.VideoCapture('video.mp4')
pTime = 0

#   
new_width = 504
new_height = 896
    
while True:
    success, img = cap.read() # img => BGR

    #    
    if not success:
        break

    #   
    img = cv2.resize(img, (new_width, new_height))

    # convert to RGB for compatibility with the mediapipe library
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)

    if results.pose_landmarks:
        mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
        for id, lm in enumerate(results.pose_landmarks.landmark):
            h, w, c = img.shape
            print(id, lm)
            cx, cy = int(lm.x * w), int(lm.y * h)
            cv2.circle(img, (cx, cy), 5, (255, 0, 0), cv2.FILLED)

    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime

    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3)
    cv2.imshow("Image", img)
    cv2.waitKey(1) # 1ms delay

#   
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import time

mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose()

cap = cv2.VideoCapture('video3.mp4')
pTime = 0

#   
new_width = 504
new_height = 896

paused = False

while True:
    if not paused:
        success, img = cap.read()  # img => BGR

        #    
        if not success:
            break

        #   
        img = cv2.resize(img, (new_width, new_height))

        # convert to RGB for compatibility with the mediapipe library
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        results = pose.process(imgRGB)
        if results.pose_landmarks:
            mpDraw.draw_landmarks(img, results.pose_landmarks, mpPose.POSE_CONNECTIONS)
            h, w, c = img.shape
            left_wrist_landmark = results.pose_landmarks.landmark[mpPose.PoseLandmark.LEFT_WRIST.value]
            cx, cy = int(left_wrist_landmark.x * w), int(left_wrist_landmark.y * h)
            if cx <= 190 and cx > 175:
                cv2.putText(img, "Warning!", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 3, (0, 0, 255), 5, cv2.LINE_AA)
            if cx <= 175:
                cv2.putText(img, "STOP!", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 3, (0, 0, 255), 5, cv2.LINE_AA)
                paused = True

        cv2.imshow("Image", img)

    key = cv2.waitKey(1) & 0xFF
    if key == ord('p'):  # 'p'    /
        paused = not paused

    if key == ord('q'):  # 'q'   
        break

    cTime = time.time()
    pTime = cTime

#   
cap.release()
cv2.destroyAllWindows()
```
